```python
from datetime import date
from dateutil.relativedelta import relativedelta

print('=' * 8 ,"LUCAS' HOTEL", '=' * 8)
quarto = []
qtde_pessoas = int(input('Quantos hóspedes estarão neste quarto?'))

for l in range(qtde_pessoas):
    print(' ' * 10, 'HÓSPEDE {}'.format(l + 1), ' ' * 10)
    print('=' * 30)
    
    nome_hospede = input('Insira o nome do {}º hóspede.'.format(l + 1)).strip().title()
    print('Nome: {}'.format(nome_hospede))
        
    dia_nascimento = input('Que dia {} nasceu?'.format(nome_hospede))
    if len(dia_nascimento) == 2 or 1:
        pass
    else:
        break
    mês_nascimento = input('Que mês {} nasceu?'.format(nome_hospede))
    if len(mês_nascimento) == 2 or 1:
        pass
    else:
        break
    ano_nascimento = input('Que ano {} nasceu?'.format(nome_hospede))
    if len(ano_nascimento) == 4:
        pass
    else:
        break

    nascimento = date(int(ano_nascimento), int(mês_nascimento), int(dia_nascimento))
    hoje = date.today()
    idade = relativedelta(hoje, nascimento)
    data_nascimento = '{}/{}/{}'.format(dia_nascimento, mês_nascimento, ano_nascimento)
    
    print('Idade: {}'.format(idade.years))
    print('Nascimento: {}'.format(data_nascimento))
    
    cpf = input('Insira o CPF')
    if len(cpf) == 14:
        cpf = '{}{}{}{}{}{}{}{}{}{}{}{}{}{}'.format(*cpf)
    elif len(cpf) == 11:
        cpf = '{}{}{}.{}{}{}.{}{}{}-{}{}'.format(*cpf)
    else:
        print("informe um CPF válido")
        break
    print('CPF: {}\n'.format(cpf))
    
    hospede = [nome_hospede, data_nascimento, idade.years, cpf ]
    quarto.append(hospede)

print(' ' * 35, 'DETALHES DO QUARTO')
print('=' * 97)

for lista in quarto:
    print('|Nome: {:<25}| Nascimento: {:^15}| Idade: {:^4}| CPF: {:^15}|'.format(lista[0], lista[1], lista[2], lista[3]))
```
